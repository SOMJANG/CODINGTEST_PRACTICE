### BaekJoon - 큰 수 (BIG) (Python) 
### 문제 : https://www.acmicpc.net/problem/14928
### 블로그 : https://somjang.tistory.com/entry/BaekJoon-14928%EB%B2%88-%ED%81%B0-%EC%88%98-BIG-Python

### Solution

```python
def BIG(N):
    return N % 20000303

if __name__ == "__main__":
    N = int(input())
    print(BIG(N))
```
